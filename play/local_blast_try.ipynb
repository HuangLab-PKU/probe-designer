```python
import os
from Bio.Blast import NCBIXML
from Bio.Blast.Applications import NcbiblastnCommandline


# Set the desired working directory
desired_directory = "/home/akikawa/GitHub/probe_designer/play/2023.5.18_sprint-seq-probe sequence"

# Change the working directory
os.chdir(desired_directory)


# Define the BLAST command line
blastn_cline = NcbiblastnCommandline(query="query.fasta", db="refseq_rna", outfmt=5, out="blast_result.xml")

# Execute the BLAST command
stdout, stderr = blastn_cline()
```

```python
# Parse the output
result_handle = open("blast_result.xml")
blast_records = NCBIXML.parse(result_handle)

# Print each alignment
for blast_record in blast_records:
    print(len(blast_record.alignments))
    for alignment in blast_record.alignments:
        for hsp in alignment.hsps:
            print('****Alignment****')
            print('sequence:', alignment.title)
            print('length:', alignment.length)
            print('score:', hsp.score)
            print('gaps:', hsp.gaps)
            print('e value:', hsp.expect)
```
